```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['REPORT_DATE'] = df['REPORT_DATE'].str.split(' ').str[0]
df['OCC_DATE'] = df['OCC_DATE'].str.split(' ').str[0]
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.drop(['OFFENCE', 'MCI_CATEGORY', 'HOOD_158', 'HOOD_140', 'NEIGHBOURHOOD_140'], axis=1, inplace=True)
```

```python
df.drop(['UCR_CODE', 'UCR_EXT'], axis=1, inplace=True)
```

```python
df.head()
```

```python
print(df['LOCATION_TYPE'].value_counts().get('Unknown', 0))
```

```python
df.drop(df[df['LOCATION_TYPE'] == 'Unknown'].index, axis=0, inplace=True)
```

```python
df.shape
```

```python
df.drop(df[df['NEIGHBOURHOOD_158'] == 'NSA'].index, axis=0, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Convert columns to desired data types
df['REPORT_DATE'] = pd.to_datetime(df['REPORT_DATE'])  # Convert 'REPORT_DATE' to datetime
df['OCC_DATE'] = pd.to_datetime(df['OCC_DATE'])  # Convert 'OCC_DATE' to datetime
df['OCC_DAY']= df['OCC_DAY'].astype(int)
df['OCC_DOY']= df['OCC_DOY'].astype(int)
df['OCC_HOUR']= df['OCC_HOUR'].astype(int)


# Display the updated data types
print(df.dtypes)
```
